```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
 
df = pd.read_csv('titanic_train.csv')
df.head()
```

> Q) Print the name of the oldest passenger travelling on ship

```python
df[df['Age'] == df['Age'].max()]['Name']
```

> Sorting

```python
df.sort_values(by=['Name','Age'],ascending=[True,False]).head()
```

> Replacing values in column

1. Through map() method

```python
d={'male':0, 'female':1}
df['Sex'] = df['Sex'].map(d)
df.head()
```

2. Through replace method

```python
o={0: 'male', 1: 'female'}
df= df.replace({'Sex':o})
df.head()
```

> Grouping Values

```python
df.groupby(by='Survived')['Age'].describe()
```

> Relation

```python
pd.crosstab(df['Sex'],df['Survived'], normalize=True)
```
