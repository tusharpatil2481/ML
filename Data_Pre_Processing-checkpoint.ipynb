```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('amsPrediction.csv')
df.head()
```

```python
df.describe()
```

> **Data Scaling using MinMaxScaler from scikit-learn**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df.values)
scaled_df = pd.DataFrame(scaled_values)
scaled_df.columns = df.columns
scaled_df.head()
```

```python
scaled_df.describe()
```

> **Data Standardization using StandardScaler from scikit-learn**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standardized_values = scaler.fit_transform(df.values)
standardized_df = pd.DataFrame(standardized_values)
standardized_df.columns = df.columns
standardized_df.head()
```

```python
standardized_df.describe()
```

### **Categorical Variables**

```python
df = pd.read_csv('stdcat.csv')
df.head()
```

> **Converting categorical data to Numeric data by replacing values**

```python
# df.Gender = df.Gender.apply(lambda x: 0 if x == 'F' else 1)
new_vals = {'F': 0, 'M': 1}
df = df.replace({ 'Gender': new_vals })
```

> **Converting categorical data to Numeric data using LabelEncoder from scikit-learn**

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['State'] = encoder.fit_transform(df['State'])
df.head()
```

> **Converting categorical data to Numeric data using OneHotEncoding from pandas dummy method**

```python
df = pd.get_dummies(df, columns=['Category'], prefix=['Cat'])
df.head()
```
